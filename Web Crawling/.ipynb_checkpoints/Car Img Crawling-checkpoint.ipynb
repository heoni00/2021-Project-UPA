# 1️⃣ Define util functions

## save_img()

```python
def save_img(url, name):
        #이미지 url과 저장할 이름을 입력하면 해당 이름으로 저장해주는 함수
        #일단 기본은 10mb까지 저장할 수 있게 해둠
    import requests
    res = requests.get(url)
    res.raise_for_status()
    img = open(name + '.png', 'wb')

    for chunk in res.iter_content(10000000): #byte로 환산해서 얼마만큼 저장해야 하는지 적으면 됨
        img.write(chunk)
    img.close()
```

## createFolder()

```python
import os

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except:
        print ('Error: Creating directory. ' +  directory)
```

# 2️⃣ HyunDai Official Web Site(by function)

```python
from tqdm import tqdm_notebook

def save_car_image_hd(root, img_name, url_base):
    i = 1
    while 1:
        if i < 10:
            url = url_base + '00' + str(i) + '.png'
        else:
            url = url_base + '0' + str(i) + '.png'
        try:
            save_img(url, root + '/' + img_name + str(i))
            print(root + '/' + img_name)
            i += 1
        except:
            break


def get_hd_car_images(img_list, dir_name='HD_IMGS'):
    BASE_DIR = os.getcwd()
    createFolder(BASE_DIR + '/' + dir_name)
    BASE_DIR += '/' + dir_name
    
    for img in tqdm_notebook(img_list):
        img_dir = BASE_DIR + '/' + img[0]
        createFolder(img_dir)
        
        save_car_image_hd(dir_name+'/'+img[0], img[0] +'_', img[1])
    
    print('Tot end')
```

```python
import requests, os, time, random
from selenium import webdriver
import bs4 as bs


class StupidSpider:
    def __init__(self, pth='chromedriver.exe'):        
        options = webdriver.ChromeOptions()
        options.add_argument('headless')
        options.add_argument('window-size=1920x1080')
        options.add_argument("user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36")
        self.driver = webdriver.Chrome(pth, options=options)
        self.driver.maximize_window()
    
    def move(self, pth):
        self.driver.get(pth)
        time.sleep(random.randrange(2,4))
    
    def get(self, xpth):
        self.target = self.driver.find_element_by_xpath(xpth)
        return self.target
    
    def gets(self, xpth):
        return self.driver.find_elements_by_xpath(xpth)
    
    def click(self, xpth="$%&"):
        if xpth != "$%&":
            tmp = self.driver.find_element_by_xpath(xpth)
            tmp.click()
            time.sleep(random.randrange(2,4))
        else:
            self.target.click()
            time.sleep(random.randrange(2,4))
    
    def send(self, key, xpth="$%&"):
        if xpth != "$%&":
            tmp = self.driver.find_element_by_xpath(xpth)
            tmp.send_keys(key)
            time.sleep(random.randrange(2,4))
        else:
            self.target.send_keys(key)
            time.sleep(random.randrange(2,4))
    
    def text(self, xpth="$%&"):
        if xpth != "$%&":
            return self.driver.find_element_by_xpath(xpth).text
        else:
            return self.target.text.replace(' ','')
    
    def get_soup(self, html = '$%&'):
        if html != '$%&':
            res = requests.get(html)
            self.soup = bs.BeautifulSoup(res.text, 'html.parser')
            return self.soup
        else:
            html = self.driver.page_source
            self.soup = bs(html, 'html.parser')
            return self.soup

    def soup_select(self, text):
        return self.soup.select(text)
    
    def get_attribute(self, text):
        return self.target.get_attribute(text)
    
    
```

```python
def tmp_url_to_page(url, page):
    return url + '"%2C"toggle"%3A%7B%7D%2C"layer"%3A""%2C"sort"%3A"ModifiedDate"%2C"page"%3A' + str(page) + '%2C"limit"%3A20%7D'
```

```python
import pandas as pd
import re #I am parsing id from url by regex
from tqdm import tqdm_notebook

spider = StupidSpider()
id_df = pd.DataFrame(columns = ['회사', '차종', '모델', 'id'])
tmp = ['', '', '', '']

for i in tqdm_notebook(range(len(url_df))):
    tmp[0] = url_df.loc[i]['회사']
    tmp[1] = url_df.loc[i]['차종']
    tmp[2] = url_df.loc[i]['모델']
    
    url = url_df.loc[i]['url']
    x = 1
# Encar switches pages by passing the page number to arg.
    while 1: 
        page = tmp_url_to_page(url, x)
        spider.move(page)
        car_list = spider.gets('//*[@id="sr_normal"]/tr')
        if len(car_list) == 0: #페이지에 차량이 없는 경우
            print(f'Escape!! at {tmp[2]} : {x} page ')
            break
        else:
            x+=1
            for k in range(1, len(car_list)+1):
                #id 얻기
                try: 
                    spider.get('//*[@id="sr_normal"]/tr[' + str(k) + ']/td[2]/a')
                except:
                    continue
                url_ = spider.target.get_attribute('href')
                id_regex = re.compile('carid=[0-9]+')
                id_ = id_regex.search(url_).group()
                tmp[3] = id_[6:]
                # 색깔 확인하는 코드인데, 자바스크립트 기반 페이지라서 그런지 계속 오류가 생긴다.
    #             spider.move('http://www.encar.com/dc/dc_cardetailview.do?pageid=dc_carsearch&listAdvType=normal&carid=' + id_)
    #             time.sleep(10)
    #             try :
    #                 spider.get('//*[@id="carPic"]/div[1]/ul[1]/li[7]')
    #             except :
    #                 spider.get('//*[@id="areaBase"]/div[2]/div[1]/div[2]/ul/li[7]')
    #                 #뷰가 두 가지라서 따로 구분함
    #             tmp[3] = spider.text()
    #             spider.move(url)

                id_df.loc[len(id_df)] = tmp
```

## Kcar

### Making url list by model

```python
#Kcar는 차종과 모델을 고르는 과정에서 ModelGrpCd와 ModelCd를 결정한 다음에 submit를 해서 페이지를 이동하는 방식
#따라서 규칙성이 명확하므로 그냥 url을 만드는 함수를 정의하는 게 편하다.
def tmp_kcar_url_gen(ModelGrpCd, ModelCd):
    return f'https://www.kcar.com/car/search/car_search_list.do?&i_sMakeCd=001&i_sModelGrpCd={ModelGrpCd}&i_sMakeModelCd=001;{ModelCd}&i_sModelCd={ModelCd}&i_sMakeType=MAKE_TYPE010'
```

```python
import pandas as pd
from tqdm import tqdm_notebook

spider = StupidSpider()
kcar_url_df = pd.DataFrame(columns = ['회사', '차종', '모델', 'url'])
ModelGrpCd = ''
ModelCd = ''

spider.move('https://www.kcar.com/index.do')

for i in tqdm_notebook(range(1,4)): # 브랜드 선택 순서대로 현대 제네시스 기아
    spider.get('//*[@id="selBrand"]')
    spider.click()
    spider.get(f'//*[@id="layerBrand"]/div/div[1]/ul/li[{i}]/button')
    tmp = ['', '', '', '']
    tmp[0] = spider.text()
    spider.click()
    
    # 모델 창을 선택
    spider.click('//*[@id="selModel"]')
    target_list = spider.gets('//*[@id="layerModel"]/ul/li/button')
    spider.click('//*[@id="selModel"]')
    for j in range(1, len(target_list)+1): 
        spider.click('//*[@id="selModel"]')
        spider.get(f'//*[@id="layerModel"]/ul/li[{j}]/button')
        
        if spider.get_attribute('data-cnt') == '0':
            continue
            
        tmp[1] = spider.text()
        ModelGrpCd = spider.get_attribute('value')
        spider.click()
        spider.click('//*[@id="selModelSub"]')
        target__list = spider.gets('//*[@id="layerModelSub"]/ul/li/button')
        for target_ in target__list:
            if target_.get_attribute('data-cnt') == '0':
                continue
            tmp[2] = target_.text
            ModelCd = target_.get_attribute('value')
            tmp[3] = tmp_kcar_url_gen(ModelGrpCd, ModelCd)
            url_df.loc[len(url_df)] = tmp   
            
kcar_url_df
```

### Making img id list by item

```python
import pandas as pd
import re #I am parsing id from url by regex
from tqdm import tqdm_notebook

spider = StupidSpider()
kcar_id_df = pd.DataFrame(columns = ['회사', '차종', '모델', 'id'])
tmp = ['', '', '', '']

for i in tqdm_notebook(range(len(url_df))):
    tmp[0] = kcar_url_df.loc[i]['회사']
    tmp[1] = kcar_url_df.loc[i]['차종']
    tmp[2] = kcar_url_df.loc[i]['모델']
    
    url = kcar_url_df.loc[i]['url']
    spider.move(url)
    spider.get('//*[@id="totalCount"]/strong')
    item_num = spider.text()
    if item_num == 0 :
        continue
    page_max = int(spider.text())/15 + 1
# Encar switches pages by passing the page number to arg.
    for page in range(1, page_max + 1):
        if page != 1: 
            page_ = page%10
            spider.click(f'//*[@id="kcarSearchListPaging"]/div/button[{page_}]')
        
        car_list = spider.gets('//*[@id="kcarSearchList"]/li/div[2]/a')
        if len(car_list) == 0: #페이지에 차량이 없는 경우
            break
            
        for k in range(1, len(car_list)+1):
            #id 얻기
            try: 
                url_ = spider.get_attribute('href')
                id_regex = re.compile('CarCd=EC[0-9]+')
                id_ = id_regex.search(url_).group()
                tmp[3] = id_[8:]
                kcar_id_df.loc[len(kcar_id_df)] = tmp
            except:
                continue
                
kcar_id_df
```
