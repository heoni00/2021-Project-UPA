## save_img()

```python
def save_img(url, name):
        #ì´ë¯¸ì§€ urlê³¼ ì €ì¥í•  ì´ë¦„ì„ ì…ë ¥í•˜ë©´ í•´ë‹¹ ì´ë¦„ìœ¼ë¡œ ì €ì¥í•´ì£¼ëŠ” í•¨ìˆ˜
        #ì¼ë‹¨ ê¸°ë³¸ì€ 10mbê¹Œì§€ ì €ì¥í•  ìˆ˜ ìˆê²Œ í•´ë‘ 
    import requests
    res = requests.get(url)
    res.raise_for_status()
    img = open(name + '.png', 'wb')

    for chunk in res.iter_content(10000000): #byteë¡œ í™˜ì‚°í•´ì„œ ì–¼ë§ˆë§Œí¼ ì €ì¥í•´ì•¼ í•˜ëŠ”ì§€ ì ìœ¼ë©´ ë¨
        img.write(chunk)
    img.close()
```

## createFolder()

```python
import os

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except:
        print ('Error: Creating directory. ' +  directory)
```

# 2ï¸âƒ£ HyunDai Official Web Site(by function)

```python
from tqdm import tqdm_notebook

def save_car_image_hd(root, img_name, url_base):
    i = 1
    while 1:
        if i < 10:
            url = url_base + '00' + str(i) + '.png'
        else:
            url = url_base + '0' + str(i) + '.png'
        try:
            save_img(url, root + '/' + img_name + str(i))
            print(root + '/' + img_name)
            i += 1
        except:
            break


def get_hd_car_images(img_list, dir_name='HD_IMGS'):
    BASE_DIR = os.getcwd()
    createFolder(BASE_DIR + '/' + dir_name)
    BASE_DIR += '/' + dir_name
    
    for img in tqdm_notebook(img_list):
        img_dir = BASE_DIR + '/' + img[0]
        createFolder(img_dir)
        
        save_car_image_hd(dir_name+'/'+img[0], img[0] +'_', img[1])
    
    print('Tot end')
```

```python
import requests, os, time, random
from selenium import webdriver
import bs4 as bs


class StupidSpider:
    def __init__(self, pth='chromedriver.exe'):        
        options = webdriver.ChromeOptions()
        options.add_argument('headless')
        options.add_argument('window-size=1920x1080')
        options.add_argument("user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36")
        self.driver = webdriver.Chrome(pth, options=options)
        self.driver.maximize_window()
    
    def move(self, pth):
        self.driver.get(pth)
        time.sleep(random.randrange(2,4))
    
    def get(self, xpth):
        self.target = self.driver.find_element_by_xpath(xpth)
        return self.target
    
    def gets(self, xpth):
        return self.driver.find_elements_by_xpath(xpth)
    
    def click(self, xpth="$%&"):
        if xpth != "$%&":
            tmp = self.driver.find_element_by_xpath(xpth)
            tmp.click()
            time.sleep(random.randrange(2,4))
        else:
            self.target.click()
            time.sleep(random.randrange(2,4))
    
    def send(self, key, xpth="$%&"):
        if xpth != "$%&":
            tmp = self.driver.find_element_by_xpath(xpth)
            tmp.send_keys(key)
            time.sleep(random.randrange(2,4))
        else:
            self.target.send_keys(key)
            time.sleep(random.randrange(2,4))
    
    def text(self, xpth="$%&"):
        if xpth != "$%&":
            return self.driver.find_element_by_xpath(xpth).text
        else:
            return self.target.text.replace(' ','')
    
    def get_soup(self, html = '$%&'):
        if html != '$%&':
            res = requests.get(html)
            self.soup = bs.BeautifulSoup(res.text, 'html.parser')
            return self.soup
        else:
            html = self.driver.page_source
            self.soup = bs(html, 'html.parser')
            return self.soup

    def soup_select(self, text):
        return self.soup.select(text)
    
    def get_attribute(self, text):
        return self.target.get_attribute(text)
    
    
```

```python
#KcarëŠ” ì°¨ì¢…ê³¼ ëª¨ë¸ì„ ê³ ë¥´ëŠ” ê³¼ì •ì—ì„œ ModelGrpCdì™€ ModelCdë¥¼ ê²°ì •í•œ ë‹¤ìŒì— submitë¥¼ í•´ì„œ í˜ì´ì§€ë¥¼ ì´ë™í•˜ëŠ” ë°©ì‹
#ë”°ë¼ì„œ ê·œì¹™ì„±ì´ ëª…í™•í•˜ë¯€ë¡œ ê·¸ëƒ¥ urlì„ ë§Œë“œëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜í•˜ëŠ” ê²Œ í¸í•˜ë‹¤.
def tmp_kcar_url_gen(ModelGrpCd, ModelCd):
    return f'https://www.kcar.com/car/search/car_search_list.do?&i_sMakeCd=001&i_sModelGrpCd={ModelGrpCd}&i_sMakeModelCd=001;{ModelCd}&i_sModelCd={ModelCd}&i_sMakeType=MAKE_TYPE010'
```

```python
import pandas as pd
from tqdm import tqdm_notebook

spider = StupidSpider()
kcar_url_df = pd.DataFrame(columns = ['íšŒì‚¬', 'ì°¨ì¢…', 'ëª¨ë¸', 'url'])
ModelGrpCd = ''
ModelCd = ''

spider.move('https://www.kcar.com/index.do')

for i in tqdm_notebook(range(1,4)): # ë¸Œëœë“œ ì„ íƒ ìˆœì„œëŒ€ë¡œ í˜„ëŒ€ ì œë„¤ì‹œìŠ¤ ê¸°ì•„
    spider.get('//*[@id="selBrand"]')
    spider.click()
    spider.get(f'//*[@id="layerBrand"]/div/div[1]/ul/li[{i}]/button')
    tmp = ['', '', '', '']
    tmp[0] = spider.text()
    spider.click()
    
    # ëª¨ë¸ ì°½ì„ ì„ íƒ
    spider.click('//*[@id="selModel"]')
    target_list = spider.gets('//*[@id="layerModel"]/ul/li/button')
    spider.click('//*[@id="selModel"]')
    for j in range(1, len(target_list)+1): 
        spider.click('//*[@id="selModel"]')
        spider.get(f'//*[@id="layerModel"]/ul/li[{j}]/button')
        
        if spider.get_attribute('data-cnt') == '0':
            continue
            
        tmp[1] = spider.text()
        ModelGrpCd = spider.get_attribute('value')
        spider.click()
        spider.click('//*[@id="selModelSub"]')
        target__list = spider.gets('//*[@id="layerModelSub"]/ul/li/button')
        for target_ in target__list:
            if target_.get_attribute('data-cnt') == '0':
                continue
            tmp[2] = target_.text
            ModelCd = target_.get_attribute('value')
            tmp[3] = tmp_kcar_url_gen(ModelGrpCd, ModelCd)
            url_df.loc[len(url_df)] = tmp   
            
kcar_url_df
```

### Making img id list by item

```python
import pandas as pd
import re #I am parsing id from url by regex
from tqdm import tqdm_notebook

spider = StupidSpider()
kcar_id_df = pd.DataFrame(columns = ['íšŒì‚¬', 'ì°¨ì¢…', 'ëª¨ë¸', 'id'])
tmp = ['', '', '', '']

for i in tqdm_notebook(range(len(url_df))):
    tmp[0] = kcar_url_df.loc[i]['íšŒì‚¬']
    tmp[1] = kcar_url_df.loc[i]['ì°¨ì¢…']
    tmp[2] = kcar_url_df.loc[i]['ëª¨ë¸']
    
    url = kcar_url_df.loc[i]['url']
    spider.move(url)
    spider.get('//*[@id="totalCount"]/strong')
    item_num = spider.text()
    if item_num == 0 :
        continue
    page_max = int(spider.text())/15 + 1
# Encar switches pages by passing the page number to arg.
    for page in range(1, page_max + 1):
        if page != 1: 
            page_ = page%10
            spider.click(f'//*[@id="kcarSearchListPaging"]/div/button[{page_}]')
        
        car_list = spider.gets('//*[@id="kcarSearchList"]/li/div[2]/a')
        if len(car_list) == 0: #í˜ì´ì§€ì— ì°¨ëŸ‰ì´ ì—†ëŠ” ê²½ìš°
            break
            
        for k in range(1, len(car_list)+1):
            #id ì–»ê¸°
            try: 
                url_ = spider.get_attribute('href')
                id_regex = re.compile('CarCd=EC[0-9]+')
                id_ = id_regex.search(url_).group()
                tmp[3] = id_[8:]
                kcar_id_df.loc[len(kcar_id_df)] = tmp
            except:
                continue
                
kcar_id_df
```
